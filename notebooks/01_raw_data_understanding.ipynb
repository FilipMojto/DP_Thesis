## Empty Methods Check

```python
# df['methods'].info()
# Count empty lists
empty_count = df['methods'].apply(lambda x: len(x) == 0).sum()

# Percentage of empty lists
empty_percent = df['methods'].apply(lambda x: len(x) == 0).mean() * 100

print(empty_count, empty_percent)
```
